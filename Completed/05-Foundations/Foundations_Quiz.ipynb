## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create x values
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# 2. Compute y = x^2
y = x**2

# 3. Plot
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Line Plot of y = x^2")
plt.grid(True)
plt.show()
```
